```python
# Bar Plot to indicate Petal Length of each Species 

plt.bar(Iris['Classification'],Iris['PL'],color='g')
plt.title('Iris Flower Species vs Petal Length')
plt.xlabel('Species')
plt.ylabel('Petal Length in cm')
```

```python
# The Specie Iris-virginica has lrgest Petal Length and Iris-setosa has smallest.
```

```python
#box plot for Sepal Length of Iris flowers

plt.boxplot(Iris['SL'])
```

```python
# the median value of Sepal length comes in between 5.5 to 6 sm 
```

```python
#Scatter Plot

sns.scatterplot(x='PL', y='PW',
                hue='Classification', data=Iris )
```

```python
#Species Setosa has smaller petal lengths and widths.
#Versicolor Species lies in the middle of the other two species in terms of petal length and width
#Species Virginica has the largest of petal lengths and widths.
```

```python
#pairplot

sns.pairplot(Iris)
```
